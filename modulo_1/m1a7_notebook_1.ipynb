# Exercício

1. Insira em uma lista todos os caminhos dos conjuntos de dados cujos títulos iniciam por "consulta_";
2. Leia todos os arquivos e insira os dataframes em uma lista;
3. Concatene todos os dataframes e reindexe as linhas.

```python
import pandas as pd
import glob
```

```python
dados_todos_estados = glob.glob('conjuntos_dados/consulta_*.csv')
print(dados_todos_estados)
```

```python
lista_df = []

for dados_estado in dados_todos_estados:
    df_estado = pd.read_csv(dados_estado, sep=';', encoding='latin_1')
    lista_df.append(df_estado)
```

```python
print(lista_df[0:1])
```

```python
df_completo = pd.concat(lista_df, axis=0, ignore_index=True)
df_completo.shape
```
