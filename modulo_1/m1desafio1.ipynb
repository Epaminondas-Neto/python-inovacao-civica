```python
import pandas as pd
```

```python
candidaturas_pe = pd.read_csv('conjuntos_dados/consulta_cand_2020_PE.csv', sep=';', encoding='latin_1')

candidaturas_pe.head()
```

```python
municipios_lista = []

for item in candidaturas_pe['NM_UE']:
    municipios_lista.append(item)
```

```python
len(municipios_lista)
```

```python
municipios_dicionario = {}

for municipio in municipios_lista:
    if municipio in municipios_dicionario:
        pass
    else:
        municipios_dicionario[municipio] = 0
```

```python
len(municipios_dicionario)
```

```python
municipios_dicionario
```

```python
for municipio in municipios_lista:
    municipios_dicionario[municipio] += 1
```
